## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(np.random.random((200,5)))
```

```python
df['date'] = pd.date_range('2023-1-1', periods=200, freq='D')
```

```python
df = df.set_index(['date'])
```

```python
print(df.loc['2023-1-1':'2023-2-10'])
```

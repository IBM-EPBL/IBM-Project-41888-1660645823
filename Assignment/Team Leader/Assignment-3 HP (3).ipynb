## Ourlier removal using IQR

```python
q1=df.Rings.quantile(0.25)  #(Q1)
q3=df.Rings.quantile(0.75)  #(Q3)
```

```python
IQR=q3-q1
```

```python
upper_limit= q3 + 1.5*IQR

lower_limit= q1 - 1.5*IQR
```

```python
df=df[df.Rings<upper_limit]
```

```python
sns.boxplot(df.Rings)
```

```python
df.shape
```
